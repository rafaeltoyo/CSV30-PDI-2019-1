## Bibliotecas

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Abrir a imagem

```python
IMG = "chromakey/6.bmp"

img = cv2.imread(IMG)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Ver resultado final
def imagem_plot(img):
    fig, axes = plt.subplots(1, 1, figsize=(25, 15), sharex=True, sharey=True)
    axes.imshow(img)
    plt.xticks([]), plt.yticks([])
    plt.show()
    
imagem_plot(img)
```

```python
try1 = img.copy()
try1 = np.float32(try1) / 255.0

try1 = ((try1 * (-0.5, 2.0, -0.5)).sum(axis=(2,)) + try1[:,:,1]).clip(min=0)

print(try1)

cv2.normalize(try1, try1, 0, 255, cv2.NORM_MINMAX)
print(try1)

try1 = np.repeat(np.int8(try1), 3, axis=1).reshape(img.shape)

imagem_plot(try1)
```
